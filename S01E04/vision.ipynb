```python
from llm import ask_llm_extract_image
import os
from openai import OpenAI
import dotenv

dotenv.load_dotenv()
client = OpenAI(
    base_url=os.getenv("BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
)
```

```python
import json
messages = [{"role": "system", "content": "Twoim zadaniem jest ekstrakcja informacji z obrazu. Interesują Cię kody tras i między jakimi miastami się one odbywają."}]
print(os.getenv("MODEL_ID"))
messages.append({
    "role": "user",
    "content": [
        {
            "type": "image_url",
            "image_url": {"url": f"data:image/png;base64,{content['data']}"}
        }
    ]
})

try:
    response = client.chat.completions.create(
        model=os.getenv("MODEL_ID"),
        messages=messages,
        # response_format={"type": "json_object"}  # Wymusza JSON od modelu
    )

    # Parsowanie odpowiedzi
    result = response.choices[0].message.content


except Exception as e:
    print(f"Błąd komunikacji z LLM: {e}")

print(result)
```
