```python
from helper import create_payload, send_payload
from pprint import pprint
import os
import json
```

```python
payload = create_payload("help")
status_code, response = send_payload(payload)
print(status_code)
print(response)
```

```python
pprint(response)

# stwórz folder data/ jeśli nie istnieje
if not os.path.exists("data"):
    os.makedirs("data")

# zapisz response w pliku json
with open("data/help.json", "w") as f:
    json.dump(response, f)
```
